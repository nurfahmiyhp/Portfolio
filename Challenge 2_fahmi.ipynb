```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sr = pd.read_csv("final-fix.csv")
```

```python
sr
```

```python
import seaborn as sns

male_population = sr.loc[sr.loc[:, 'sex']=='male',:]
female_population = sr.loc[sr.loc[:, 'sex']=='female',:]

plt.figure(figsize=(8,4))

lp_male = sns.lineplot(x = 'year' , y = 'suicides_no' , data = male_population)
lp_female = sns.lineplot(x = 'year' , y = 'suicides_no' , data = female_population)
leg1 = plt.legend(['Males','Females'])
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x = "country", y = "suicides_no", hue = "age", data = sr)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,5))
sns.scatterplot(x = "country", y = "suicides_no", hue = "age", data = sr, estimator=max)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,5))
sns.scatterplot(x = "country", y = "suicides_no", hue = "age", data = sr, estimator=max)
plt.xticks(rotation=90)
plt.legend(['5-14 years','15-24 years','25-34 years','35-54 years','55-74 years','75+ years'])
plt.show()
```
